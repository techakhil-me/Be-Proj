```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class LungCancerDetectionPipeline:
    def __init__(self):
        self.img_height = 224
        self.img_width = 224
        self.batch_size = 32
        self.classes = ['Bengin', 'Malignant', 'Normal']  # Matches dataset folder names
        self.model = None
        
    def prepare_data(self, base_dir):
        """Prepare and augment the image data"""
        print("Preparing data...")
        
        # Data augmentation for training
        self.train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            validation_split=0.2
        )

        # Only rescaling for validation
        self.valid_datagen = ImageDataGenerator(
            rescale=1./255,
            validation_split=0.2
        )

        print("Loading training data...")
        self.train_generator = self.train_datagen.flow_from_directory(
            base_dir,
            target_size=(self.img_height, self.img_width),
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='training',
            shuffle=True
        )

        print("Loading validation data...")
        self.validation_generator = self.valid_datagen.flow_from_directory(
            base_dir,
            target_size=(self.img_height, self.img_width),
            batch_size=self.batch_size,
            class_mode='categorical',
            subset='validation',
            shuffle=True
        )
        
        print("Data preparation completed!")

    def build_model(self):
        """Build the CNN model architecture"""
        print("Building model...")
        
        self.model = models.Sequential([
            # First Convolutional Block
            layers.Conv2D(32, 3, activation='relu', padding='same', 
                         input_shape=(self.img_height, self.img_width, 3)),
            layers.BatchNormalization(),
            layers.Conv2D(32, 3, activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(),
            layers.Dropout(0.25),
            
            # Second Convolutional Block
            layers.Conv2D(64, 3, activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.Conv2D(64, 3, activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(),
            layers.Dropout(0.25),
            
            # Third Convolutional Block
            layers.Conv2D(128, 3, activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.Conv2D(128, 3, activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(),
            layers.Dropout(0.25),
            
            # Dense Layers
            layers.Flatten(),
            layers.Dense(512, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(len(self.classes), activation='softmax')
        ])
        
        # Compile the model
        self.model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        print("Model built successfully!")
        self.model.summary()
        
    def train_model(self, epochs=50):
        """Train the model with early stopping"""
        print("Starting model training...")
        
        # Create models directory
        os.makedirs('models', exist_ok=True)
        
        # Callbacks
        early_stopping = tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )
        
        checkpoint = tf.keras.callbacks.ModelCheckpoint(
            'models/best_model.keras',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max'
        )
        
        # Reduce learning rate when plateau is reached
        reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=1e-6
        )
        
        # Train the model
        self.history = self.model.fit(
            self.train_generator,
            epochs=epochs,
            validation_data=self.validation_generator,
            callbacks=[early_stopping, checkpoint, reduce_lr]
        )
        
        # Save final model
        self.model.save('models/final_model.keras')
        print("Training completed!")
        
    def plot_training_history(self):
        """Plot training history"""
        plt.figure(figsize=(12, 4))
        
        # Plot accuracy
        plt.subplot(1, 2, 1)
        plt.plot(self.history.history['accuracy'], label='Training Accuracy')
        plt.plot(self.history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        
        # Plot loss
        plt.subplot(1, 2, 2)
        plt.plot(self.history.history['loss'], label='Training Loss')
        plt.plot(self.history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('training_history.png')
        plt.show()
        
    def evaluate_model(self):
        """Evaluate the model and generate performance metrics"""
        print("\nEvaluating model...")
        
        # Get predictions for validation set
        predictions = []
        true_labels = []
        
        for i in range(len(self.validation_generator)):
            x, y = self.validation_generator[i]
            pred = self.model.predict(x)
            pred_classes = np.argmax(pred, axis=1)
            true_classes = np.argmax(y, axis=1)
            
            predictions.extend(pred_classes)
            true_labels.extend(true_classes)
            
            if len(predictions) >= len(self.validation_generator.labels):
                break
                
        # Generate classification report
        print("\nClassification Report:")
        print(classification_report(true_labels[:len(self.validation_generator.labels)],
                                 predictions[:len(self.validation_generator.labels)],
                                 target_names=self.classes))
        
        # Generate confusion matrix
        cm = confusion_matrix(true_labels[:len(self.validation_generator.labels)],
                            predictions[:len(self.validation_generator.labels)])
        
        # Plot confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=self.classes,
                   yticklabels=self.classes)
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.savefig('confusion_matrix.png')
        plt.show()

def main():
    # Dataset path - update this to your dataset location
    dataset_path = "/kaggle/input/iqothnccd-lung-cancer-dataset/The IQ-OTHNCCD lung cancer dataset/The IQ-OTHNCCD lung cancer dataset"  # This should be the folder containing Bengin, Malignant, and Normal folders
    
    # Initialize pipeline
    print("Initializing pipeline...")
    pipeline = LungCancerDetectionPipeline()
    
    # Check if dataset exists
    if not os.path.exists(dataset_path):
        raise FileNotFoundError(f"Dataset not found at {dataset_path}")
    
    # Print dataset structure
    print("\nDataset structure:")
    for root, dirs, files in os.walk(dataset_path):
        level = root.replace(dataset_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        if level == 1:
            print(f"{indent}    {len(files)} images")
    
    # Prepare data
    pipeline.prepare_data(dataset_path)
    
    # Build model
    pipeline.build_model()
    
    # Train model
    pipeline.train_model(epochs=50)
    
    # Plot training history
    pipeline.plot_training_history()
    
    # Evaluate model
    pipeline.evaluate_model()
    
    print("\nTraining and evaluation completed!")
    print("Saved files:")
    print("- models/best_model.keras (Best model weights)")
    print("- models/final_model.keras (Final model weights)")
    print("- training_history.png (Training plots)")
    print("- confusion_matrix.png (Evaluation results)")

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print(f"\nError occurred: {str(e)}")
        raise
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import Image

# Set dataset path
dataset_path = "/kaggle/input/iqothnccd-lung-cancer-dataset/The IQ-OTHNCCD lung cancer dataset/The IQ-OTHNCCD lung cancer dataset"

# Define classes
classes = ['Bengin cases', 'Malignant cases', 'Normal cases']

def display_dataset_statistics():
    """Display statistics of dataset images"""
    print("\nDataset structure and image counts:")

    # Count images per class
    class_counts = {}
    total_images = 0

    for class_name in classes:
        class_path = os.path.join(dataset_path, class_name)
        if os.path.exists(class_path):
            num_images = len(os.listdir(class_path))
            class_counts[class_name] = num_images
            total_images += num_images
        else:
            print(f"Warning: {class_name} directory not found.")

    # Display statistics
    for class_name, count in class_counts.items():
        print(f"{class_name}: {count} images")

    print(f"\nTotal Images: {total_images}")

    # Plot bar chart
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()), palette="coolwarm")
    plt.title("Lung Cancer Dataset Image Distribution")
    plt.ylabel("Number of Images")
    plt.xlabel("Class")
    plt.show()

def preprocess_images():
    """Preprocess dataset images and visualize augmentation"""
    print("\nPreprocessing images...")

    # Image preprocessing settings
    img_height, img_width = 224, 224

    # Data augmentation setup
    datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True
    )

    # Find a valid sample image
    sample_image_path = None
    for class_name in classes:
        class_path = os.path.join(dataset_path, class_name)
        images = os.listdir(class_path) if os.path.exists(class_path) else []
        if images:
            sample_image_path = os.path.join(class_path, images[0])
            break

    if sample_image_path is None:
        raise ValueError("No images found across all classes!")

    print(f"Using sample image: {sample_image_path}")
    
    # Load and preprocess the image
    sample_image = Image.open(sample_image_path).resize((img_width, img_height))
    sample_array = np.array(sample_image) / 255.0
    sample_array = np.expand_dims(sample_array, axis=0)  # Add batch dimension

    # Apply data augmentation
    augmented_images = [sample_array[0]]  # Store original image without batch dimension
    for i in range(4):  # Generate 4 augmented samples
        aug_img = datagen.flow(sample_array, batch_size=1)[0][0]  # Remove batch dim
        augmented_images.append(aug_img)

    # Display original and augmented images
    plt.figure(figsize=(10, 5))
    for i, img in enumerate(augmented_images):
        plt.subplot(1, 5, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title("Original" if i == 0 else f"Aug {i}")
    
    plt.show()
    print("Image preprocessing completed!")

# Run the function
preprocess_images()
```
